## Sequential Chain

```python
hf_token = os.getenv("HF_TOKEN")
```

```python
llm_endpoint = HuggingFaceEndpoint(
    repo_id="deepseek-ai/DeepSeek-R1",
    task = 'text-generation',
    temperature=0.9,
    huggingfacehub_api_token=hf_token)
```

```python
model = ChatHuggingFace(llm=llm_endpoint) 
```

```python
templet1 = PromptTemplate(template="Get a detailed report on the following topic: {topic}",
                          input_variables=['topic'])

templte2 = PromptTemplate(template="Create a 5 point summary of the follwoing report {report}",
                          input_variables=['report'])
```

```python
parser = StrOutputParser()
```

```python
chain = templet1 | model | parser | templte2 | model | parser
```

```python
response = chain.invoke({'topic':'LLM vs AI agent'})
print(response)
```

```python
chain.get_graph().print_ascii()
```

## Parallel Chains

```python
from langchain_classic.schema.runnable import RunnableParallel
```

```python
google_api_key = os.getenv('GOOGLE_API_KEY1')
```

```python
model2 = ChatGoogleGenerativeAI(model='gemini-2.5-flash-lite',
                                api_key= google_api_key)
```

```python
templet1 = PromptTemplate(template="Get 10 latest hard interview questions along with small answers on the following topic: {topic}",
                          input_variables=['topic'])

templet2 = PromptTemplate(template="Get 10 latest hard multiple choice questions along with correct answers on the following topic: {topic}",
                          input_variables=['topic'])

templet3 = PromptTemplate(template="Merge these two texts in one: {Q&A} \n {MCQ}",
                          input_variables=['Q&A','MCQ'])
```

```python
parallel_chain = RunnableParallel({
    'Q&A': templet1 | model2 | parser,
    'MCQ': templet2 | model | parser
})

merger = templet3 | model | parser

chain = parallel_chain | merger
```

```python
response = chain.invoke({'topic':'Generative AI'})
```

```python
print(response)
```

## Conditional Chains

```python
from langchain_classic.schema.runnable import RunnableBranch, RunnableLambda
from langchain_classic.output_parsers import PydanticOutputParser
from pydantic import BaseModel, Field
from typing import Literal
```

```python
class Description(BaseModel):
    description : Literal['long','short'] = Field(description='Classify long if user wants detailed report else short')
    
```

```python
parser_pydantic = PydanticOutputParser(pydantic_object=Description)
```

```python
templet1 = PromptTemplate(template='''Understand the following text given by the 
                          user and classify if user want detailed report or small 
                          dedcription/ summary. Here is the text : {text}.
                          use the following output structure {structure}''',
                          input_variables=['text'],
                          partial_variables={'structure':parser_pydantic.get_format_instructions()})

templet2 = PromptTemplate(template='Generate the detailed report on the topic : {text}',
                          input_variables=['text'])

templet3 = PromptTemplate(template='Generate a short summary on the topic : {text}',
                          input_variables=['text'])
```

```python
input_chain = templet1 | model | parser_pydantic
```

```python
cond_chain = RunnableBranch(
    (lambda x:x.description=='short', templet3 | model | parser),
    (lambda x:x.description=='long', templet2 | model2 | parser),
    RunnableLambda(lambda x:"could not classify text")
)
```

```python
chain =  input_chain | cond_chain
```

```python
response = chain.invoke({'text':"Brief about transfomer architecture"})
```
