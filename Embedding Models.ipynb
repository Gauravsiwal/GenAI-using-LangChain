```python
# Closed Source Embedding Models

from langchain_openai import OpenAIEmbeddings
from google.colab import userdata
```

```python
openai_key = userdata.get('OPENAI_API_KEY')
```

```python
embed_model = OpenAIEmbeddings(model='text-embedding-3-small',dimensions=64,
                               api_key=openai_key)
```

```python
sentance = 'What are the assumptions of linear regression model?'
```

```python
vector_embed = embed_model.embed_query(sentance)
```

```python
len(vector_embed)
```

```python
# Embedding of documents

doc = ['Residual must be normal',
       'New Delhi is capital of India',
       'Bias and variance must be low']
```

```python
doc_vectors = embed_model.embed_documents(doc)
```

```python
for n,i in enumerate(doc_vectors):
  print(f'Doc : {n+1}',i)
```
