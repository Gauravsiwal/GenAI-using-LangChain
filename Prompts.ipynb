Attention Is All You Need" introduces the Transformer, a novel neural network architecture that **dispenses with recurrent and convolutional layers.**  Its core innovation is the **self-attention mechanism,** allowing it to **weigh the importance of different input tokens** regardless of their position. This enables the model to **capture long-range dependencies** more effectively than previous models.  The Transformer consists of an encoder-decoder structure where **each layer heavily relies on multi-head self-attention and feed-forward networks.**  This architecture achieves **state-of-the-art results** on machine translation tasks while being **more parallelizable** and requiring **less training time.** The paper argues that attention alone is a sufficient mechanism for many sequence transduction problems, **paving the way for large pre-trained language models.**

## **Dynamic Prompt**

```python
templet = PromptTemplate(template="""
<Role> You are an expert in summarizing research papers.
<Goal> Summarize the following research paper {paper_name}.
<Constraints>
1. The style of the summary should be :{summary_style}.
2. The length of the summary should be: {length}.
3. Do not add any new information. Incase if some thing is not available simply
display 'Information not available'
""",
                         input_variables=['paper_name','summary_style','length'],
                         validate_template=True)
```

```python
paper_name  = input('Enter the paper name.')
style = input('Enter the summary style.')
length = input('Enter the length of summary.')
```

```python
prompt = templet.invoke({'paper_name':paper_name,
                         'summary_style':style,
                         'length':length})
```

```python
response = llm_model.invoke(prompt)
```

```python
print(response.text)
```

The paper "Attention Is All You Need" introduces the Transformer architecture, a novel neural network that dispenses with recurrence and convolutions entirely, relying solely on attention mechanisms.

The core of the Transformer is the self-attention mechanism. For an input sequence $X = (x_1, \dots, x_n)$, the self-attention function computes a representation of each element $x_i$ as a weighted sum of all elements in the sequence. This is achieved by first computing three vectors for each input: Query ($Q$), Key ($K$), and Value ($V$), which are linear projections of the input embeddings. The attention score for a pair of elements $(x_i, x_j)$ is calculated as a scaled dot product of the query vector for $x_i$ and the key vector for $x_j$: $score(Q_i, K_j) = \frac{Q_i \cdot K_j}{\sqrt{d_k}}$. These scores are then normalized using a softmax function to obtain attention weights $\alpha_{ij} = \frac{\exp(score(Q_i, K_j))}{\sum_{k=1}^n \exp(score(Q_i, K_k))}$. The output for $x_i$ is then computed as the weighted sum of the value vectors: $Attention(Q_i, K, V) = \sum_{j=1}^n \alpha_{ij} V_j$.

Multi-Head Attention extends this by running the attention mechanism in parallel with different, learned linear projections of the queries, keys, and values. The outputs from these multiple "heads" are concatenated and linearly projected. This allows the model to jointly attend to information from different representation subspaces at different positions.

The Transformer encoder consists of a stack of identical layers, each containing a multi-head self-attention mechanism followed by a simple, position-wise fully connected feed-forward network. The decoder also has an encoder-decoder attention mechanism, allowing it to attend over the output of the encoder. Positional encodings are added to the input embeddings to inject information about the relative or absolute position of tokens. The feed-forward network in each layer applies the same linear transformation at each position, followed by a ReLU activation: $FFN(x) = \max(0, xW_1 + b_1)W_2 + b_2$. The entire model is trained with the standard cross-entropy loss.

**Attention is All You Need: A Beginner-Friendly Summary**

Imagine you're trying to understand a really long sentence or a story. Traditionally, computers would try to read it word by word, in order. This is like a person reading a book very slowly, focusing on each word before moving to the next.

The "Attention is All You Need" paper introduced a completely new way for computers to understand language, called the **Transformer**. The key idea here is "attention."

Instead of just reading word by word, the Transformer model can look at *all* the words in a sentence (or even a paragraph) at the same time. Then, it figures out which words are most important or related to each other. Think of it like highlighting the key phrases in a text that help you understand the main idea.

The "attention mechanism" is like a spotlight that the model shines on different words to see how they connect. For example, in the sentence "The cat sat on the mat," when the model processes "sat," it can "pay attention" to both "cat" (who is doing the sitting) and "mat" (where the sitting is happening).

This is much faster and more effective than older methods. It allows the computer to grasp the overall meaning and context of the language much better. This powerful technique has become the foundation for many amazing language AI systems we see today, like chatbots that can write stories or translate languages.

## **List of Messages**

* using message class in langchain

**Lets create a simple chatbot to demonstrate**

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
```

```python
chat_history = [SystemMessage(content='You are a helpful assistant')]
```

```python
while True:
  user_input = input('You:')
  chat_history.append(HumanMessage(content=user_input))
  if user_input.lower() == 'exit':
    break
  response = llm_model.invoke(chat_history)
  print('AI:',response.content)
  chat_history.append(AIMessage(content=response.content))
```

## **Dynamic Prompting in chat messages**

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
chat_template = ChatPromptTemplate([
    ('system','You are an expert in the field of {domain}'),
    ('human','explain in simple terms: {topic}')])
```

```python
chat_history = chat_template.invoke({'domain':'Human resources',
                      'topic':'How to negotiate salary while new recruitment'}).messages
```

```python
response = llm_model.invoke(chat_history)
```

```python
chat_history.append(AIMessage(content=response.content))
```
