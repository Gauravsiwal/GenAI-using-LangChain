```python
from langchain_google_genai import ChatGoogleGenerativeAI
import os
from typing import TypedDict, Annotated, Literal, Optional
```

```python
api_key = os.getenv("GOOGLE_API_KEY1")
model = ChatGoogleGenerativeAI(model='gemini-2.5-flash-lite',
                               api_key= api_key)
```

```python
class Answer(TypedDict):
    country : Annotated[str,'The name of country itself']
    capital : Annotated[str,'The capital of country']
    currency : Annotated[Optional[str],'The currency of that country']
    English : Annotated[Literal['Yes','No'],'Does this country speaks english']
```

```python
structured_model = model.with_structured_output(Answer)
```

```python
response = structured_model.invoke('Give information on Uzbekistan')
```

```python
response
```

## Pydantic

```python
from pydantic import BaseModel, Field
```

```python
class Output_Structure(BaseModel):
    country : str = Field(description='Name of the country')
    capital : str = Field(description='Capital of the country')
    currency : str = Field(description='Currency of the country')
    English : Literal['Yes','No'] = Field(description='Do majority people speak english in this country?')
```

```python
structured_model = model.with_structured_output(Output_Structure)
```

```python
response = structured_model.invoke('Give details of combodia')
```

```python
response.model_dump()
```

```python
response.model_dump_json()
```

```python
type(response.model_dump())
```

## JSON

```python
json_schema = {
  "type": "object",
  "properties": {
    "country": {
      "type": "string",
      "description": "Name of the country"
    },
    "capital": {
      "type": "string",
      "description": "Capital of the country"
    },
    "currency": {
      "type": "string",
      "description": "Currency of the country"
    },
    "English": {
      "type": "string",
      "enum": ["Yes", "No"],
      "description": "Do majority people speak english in this country?"
    }
  },
  "required": [
    "country",
    "capital",
    "currency",
    "English"
  ]
}
```

```python
structured_model = model.with_structured_output(json_schema)
```

```python
response = structured_model.invoke('Give details of combodia')
```
