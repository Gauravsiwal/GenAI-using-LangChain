# API Based Retrievers

## Wikipedia Retriever

```python
query = 'Change in US politics after making epstein files public'
```

```python
retriever = WikipediaRetriever(top_k_results=2)
```

```python
docs = retriever.invoke(query)
```

```python
for i,doc in enumerate(docs):
  print('*'*20,f'Result {i+1}','*'*20)
  print(doc.page_content)
```

## Tavily Search Retriever

```python
!pip install tavily -q
```

```python
import os
```

```python
os.environ['TAVILY_API_KEY'] = userdata.get('TAVILY_API_KEY')
```

```python
query = 'Change in US politics after making epstein files public'

retriever = TavilySearchAPIRetriever(k=3)

docs = retriever.invoke(query)

for doc in docs:
    print(doc.page_content)
```

## ArXIV Retriever

```python
!pip install arxiv -q
```

```python
query = "Difference between encoder and decoder transformer architecture?"

retriever = ArxivRetriever(
    load_max_docs=2,
    get_full_documents=False,
)

docs = retriever.invoke(query)

for doc in docs:
    print(doc.page_content)
```
