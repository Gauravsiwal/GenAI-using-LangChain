### OpenAI

```python
from langchain_openai import OpenAI, ChatOpenAI
from google.colab import userdata
```

```python
## OpenAI LLM
```

```python
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')

prompt = "What is linear regression in machine learning?"

llm = OpenAI(model = "gpt-4o-mini",api_key=OPENAI_API_KEY)
```

```python
response = llm.invoke(prompt)
```

```python
print(response)
```

```python
## OpenAI: Chatmodel
```

```python
model = ChatOpenAI(model = "gpt-4o-mini",
                   api_key=OPENAI_API_KEY,
                   temperature=1.8,
                   max_completion_tokens=15)
```

```python
prompt = "Tell me one joke on charted accountants"
response = model.invoke(prompt)
```

```python
print(response.content)
```

## Language Models : Open Source Models

**Can be accessed from Hugging Face Library**

```python
pip install httpx==0.27.0
```

```python
# Using hugging gace inference API
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
```

```python
hf_token_access = userdata.get('HF_ACCESS_TOKEN')
```

```python
llm_endpoint = HuggingFaceEndpoint(
    repo_id="deepseek-ai/DeepSeek-R1",
    task = 'text-generation',
    temperature=0.9,
    huggingfacehub_api_token=hf_token_access,
    max_new_tokens=1000
)
```

```python
model = ChatHuggingFace(llm=llm_endpoint)
```

```python
response = model.invoke('Explain Logistic regrssion')
```

```python
print(response.content)
```
